```python
import time

def latency(func, *args):
    start = time.perf_counter()
    result = func(*args)
    end = time.perf_counter()
    return result, end - start

x = lambda n: "even" if n % 2 == 0 else "odd"

output, time_taken = latency(x, 50)

print("Output:", output)
print("Latency:", time_taken, "seconds")
```
